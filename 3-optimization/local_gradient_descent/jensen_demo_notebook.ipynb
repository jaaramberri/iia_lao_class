# Jensen's Inequality Demo

We illustrate that for a convex function $f(x)=1/x$:

$$\frac{1}{\mathbb{E}[X]} < \mathbb{E}\left[\frac{1}{X}\right]$$


```python
import numpy as np
import matplotlib.pyplot as plt

# Fix seed
np.random.seed(4)

# Generate data
X = 3 + 0.7 * np.random.randn(12)
X = np.clip(X, 0.4, None)

# Compute quantities
mean_X = np.mean(X)
inv_mean = 1 / mean_X
mean_inv = np.mean(1 / X)

print("E[X]      =", mean_X)
print("1/E[X]    =", inv_mean)
print("E[1/X]    =", mean_inv)
print()
print(f"1/E[X] = {inv_mean:.6f}  <  E[1/X] = {mean_inv:.6f}")
print("Inequality holds:", inv_mean < mean_inv)
```

```python
# Plot
x_plot = np.linspace(0.4, 5.0, 400)
y_plot = 1 / x_plot

plt.figure(figsize=(10,6))

plt.plot(x_plot, y_plot, linewidth=2.5, label=r"$f(x)=1/x$")

plt.scatter(X, 1/X, s=90, alpha=0.85, label="Samples")

plt.scatter(mean_X, inv_mean, s=180, marker="X", label=r"$(E[X],1/E[X])$")
plt.scatter(mean_X, mean_inv, s=180, marker="o", label=r"$(E[X],E[1/X])$")

plt.plot([mean_X, mean_X], [inv_mean, mean_inv], '--', linewidth=2)

plt.axvline(mean_X, linestyle=':', linewidth=1.8, label=r"$E[X]$")

plt.xlabel("x")
plt.ylabel("1/x")
plt.title("Jensen Inequality Demo")

plt.grid(True, alpha=0.3)
plt.legend()

plt.show()
```
