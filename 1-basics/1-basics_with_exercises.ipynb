### 1. Scalars, vectors, matrices, and tensors

A **scalar** is a single number.  
A **vector** is an ordered list of numbers.  
A **matrix** is a rectangular array of numbers.  
A **tensor** is a multidimensional array.

In AI:
- scalars can represent a loss value,
- vectors can represent features or embeddings,
- matrices can represent weight layers,
- tensors can represent batches of images, sequences, or activations.


```python
scalar = 3.14
vector = np.array([1, 2, 3])
matrix = np.array([[1, 2], [3, 4]])
tensor = np.arange(24).reshape(2, 3, 4)

print("Scalar:", scalar)
print("Vector shape:", vector.shape)
print("Matrix shape:", matrix.shape)
print("Tensor shape:", tensor.shape)
```

### 2. Matrix-vector and matrix-matrix multiplication

If $A \in \mathbb{R}^{m \times n}$ and $\mathbf{x} \in \mathbb{R}^n$, then $A\mathbf{x} \in \mathbb{R}^m$.

If $A \in \mathbb{R}^{m \times n}$ and $B \in \mathbb{R}^{n \times p}$, then $AB \in \mathbb{R}^{m \times p}$.

In machine learning, matrix-vector multiplication appears in the forward pass of linear layers.


```python
A = np.array([[1, 2],
              [3, 4]])
x = np.array([1, -1])
B = np.array([[2, 0],
              [1, 5]])

print("A @ x =", A @ x)
print("A @ B =\n", A @ B)
```

### 3. Matrices as linear transformations

A matrix can be viewed as a transformation that sends an input vector to an output vector.

Examples:
- scaling,
- rotation,
- reflection,
- projection.

This viewpoint is very useful in AI because weight matrices transform feature vectors into new representations.


```python
theta = np.pi / 6  # 30 degrees
R = np.array([
    [np.cos(theta), -np.sin(theta)],
    [np.sin(theta),  np.cos(theta)]
])

v = np.array([1.0, 0.0])
Rv = R @ v

print("Original vector:", v)
print("Rotated vector:", Rv)
```

### 4. Identity and inverse matrices

The **identity matrix** leaves vectors unchanged:
$ I\mathbf{x} = \mathbf{x} $.

If $A$ is invertible, then there exists $A^{-1}$ such that
$ AA^{-1} = A^{-1}A = I $.

In practice, we often solve linear systems instead of explicitly computing inverses.


```python
A = np.array([[2.0, 1.0],
              [1.0, 1.0]])
I = np.eye(2)
A_inv = np.linalg.inv(A)

print("I =\n", I)
print("A^{-1} =\n", A_inv)
print("A @ A^{-1} =\n", A @ A_inv)
```

### 5. Linear systems, dependence, and span

A **linear system** has the form $A\mathbf{x} = \mathbf{b}$.

Vectors are **linearly dependent** if one can be written as a combination of the others.

The **span** of vectors is the set of all linear combinations of them.

In AI, span and dependence matter in embeddings, feature spaces, and dimensionality reduction.


```python
x1 = np.array([1, 2, 3])
x2 = np.array([2, 4, 6])
x3 = np.array([1, 0, 1])

print("x2 == 2*x1 ?", np.allclose(x2, 2*x1))

A = np.column_stack([x1, x3])
coeffs = np.array([3, -1])
y = A @ coeffs

print("A =\n", A)
print("Example vector in the span:", y)
```

### 6. Special kinds of matrices

Some important classes of matrices:
- **diagonal**: only diagonal entries may be nonzero,
- **symmetric**: $A = A^T$,
- **orthogonal**: $Q^TQ = I$.

Symmetric matrices appear in covariance matrices and Hessians. Orthogonal matrices preserve lengths and angles.


```python
D = np.diag([1, 2, 3])
S = np.array([[2, 1],
              [1, 2]])
Q = np.array([[0, -1],
              [1,  0]])  # rotation by 90°

print("Diagonal matrix D =\n", D)
print("S symmetric? ", np.allclose(S, S.T))
print("Q orthogonal? ", np.allclose(Q.T @ Q, np.eye(2)))
```

### 7. Dot product

For vectors $\mathbf{u}, \mathbf{v} \in \mathbb{R}^n$, the dot product is

$ \mathbf{u} \cdot \mathbf{v} = \sum_i u_i v_i $.

It measures alignment between vectors and is used in similarity scores and attention mechanisms.


```python
u = np.array([1, 3, -5])
v = np.array([4, -2, -1])

dot = u @ v
norm_u = np.linalg.norm(u)
norm_v = np.linalg.norm(v)
cos_alpha = dot / (norm_u * norm_v)

print("u · v =", dot)
print("||u|| =", norm_u)
print("||v|| =", norm_v)
print("cos(alpha) =", cos_alpha)
```

### 8. Inner product

An **inner product** generalizes the dot product. In finite-dimensional Euclidean space, the standard inner product is just the dot product.

For functions, one common inner product is

$ \displaystyle \langle u, v \rangle = \int_\Omega u(x)v(x)\,dx $.

This idea appears in functional analysis, kernels, and continuous models.


```python
x = np.linspace(0, 1, 1000)
u = x
v = x**2

inner_product = np.trapz(u * v, x)
print("Approximate function inner product <u,v> =", inner_product)
```

### 9. Norms and $L^p$ norms

A **norm** measures the size of a vector.

For $p \ge 1$, the $L^p$ norm is

$\displaystyle \|\mathbf{x}\|_p = \left(\sum_i |x_i|^p\right)^{1/p} $.

Norms are used in regularization, optimization, and distance calculations.


```python
x = np.array([1.0, -2.0, 3.0])

for p in [1, 2, 3]:
    val = np.sum(np.abs(x)**p)**(1/p)
    print(f"L^{p} norm =", val)
```

### 10. $L^2$, $L^1$, and other norms

- The **$L^2$ norm** is the Euclidean norm.
- The **$L^1$ norm** sums absolute values and is linked to sparsity.
- The **$L^\infty$ norm** takes the largest absolute entry.
- The **Frobenius norm** extends $L^2$-type measurement to matrices.


```python
x = np.array([1.0, -2.0, 3.0])
A = np.array([[1, 2],
              [3, 4]])

l1 = np.linalg.norm(x, 1)
l2 = np.linalg.norm(x, 2)
linf = np.linalg.norm(x, np.inf)
fro = np.linalg.norm(A, "fro")

print("L1 norm =", l1)
print("L2 norm =", l2)
print("L-infinity norm =", linf)
print("Frobenius norm =", fro)
```

### 11. Eigenvalues and eigenvectors

An eigenvector of a matrix $A$ is a nonzero vector $\mathbf{x}$ such that

$ A\mathbf{x} = \lambda \mathbf{x} $.

The vector keeps its direction up to scaling, and $\lambda$ is the scaling factor.

Eigenvalues and eigenvectors are central in PCA and covariance analysis.


```python
A = np.array([[2, 1],
              [1, 2]])

eigvals, eigvecs = np.linalg.eig(A)

print("Eigenvalues:", eigvals)
print("Eigenvectors (columns):\n", eigvecs)
```

### 12. Eigendecomposition

If a matrix has enough linearly independent eigenvectors, it can be decomposed as

$ A = Q\Lambda Q^{-1} $,

where $Q$ contains eigenvectors and $\Lambda$ contains eigenvalues.

For symmetric matrices, the decomposition is especially nice because $Q$ can be orthogonal.


```python
A = np.array([[2, 1],
              [1, 2]])

eigvals, Q = np.linalg.eig(A)
Lambda = np.diag(eigvals)
A_reconstructed = Q @ Lambda @ np.linalg.inv(Q)

print("Q =\n", Q)
print("Lambda =\n", Lambda)
print("Q Lambda Q^{-1} =\n", A_reconstructed)
```

### 13. Singular Value Decomposition (SVD)

Any matrix $A$ can be written as

$ A = U\Sigma V^T $.

SVD is fundamental in:
- dimensionality reduction,
- compression,
- recommender systems,
- numerical stability,
- PCA-related computations.


```python
A = np.array([[1, 2],
              [3, 4],
              [5, 6]], dtype=float)

U, s, VT = np.linalg.svd(A, full_matrices=False)
Sigma = np.diag(s)
A_reconstructed = U @ Sigma @ VT

print("Singular values:", s)
print("U =\n", U)
print("Sigma =\n", Sigma)
print("V^T =\n", VT)
print("Reconstruction error =", np.linalg.norm(A - A_reconstructed))
```

### 14. Derivative

The derivative measures the instantaneous rate of change of a function and gives the best local linear approximation.

For a function $f(x)$, the derivative at $x_0$ is

$\displaystyle  f'(x_0) = \lim_{h\to 0} \frac{f(x_0+h)-f(x_0)}{h} $.


```python
# ======================================================
# Function
# ======================================================

def f(x):
    return np.log(1 + np.exp(2*x))

# Point where we evaluate the derivative
x0 = 0.5

# ======================================================
# 1. Numerical derivative (finite differences)
# ======================================================

h = 1e-5
numerical_derivative = (f(x0 + h) - f(x0)) / h

# ======================================================
# 2. Exact derivative written by hand
# ======================================================

exact_derivative = 2*np.exp(2*x0) / (1 + np.exp(2*x0))

# ======================================================
# 3. Symbolic derivative with SymPy
# ======================================================

x = sp.symbols('x')
f_sym = sp.log(1 + sp.exp(2*x))

df_sym = sp.diff(f_sym, x)             # symbolic derivative
df_sym_simplified = sp.simplify(df_sym)

df_sym_func = sp.lambdify(x, df_sym_simplified, "numpy")
symbolic_derivative = df_sym_func(x0)

# ======================================================
# Results
# ======================================================

print("Function:")
print("f(x) = log(1 + exp(2x))")
print()

print("Symbolic derivative:")
print(df_sym_simplified)
print()

print("Numerical derivative:", numerical_derivative)
print("Exact derivative:    ", exact_derivative)
print("Symbolic derivative: ", symbolic_derivative)
print()

print("Error (numerical vs exact):   ", abs(numerical_derivative - exact_derivative))
print("Error (symbolic vs exact):    ", abs(symbolic_derivative - exact_derivative))
```

### 15. Gradient and directional derivative

For a scalar function of several variables, the **gradient** collects all first partial derivatives.

The gradient points in the direction of steepest ascent, so $-\nabla f$ gives a basic descent direction.

The **directional derivative** measures the rate of change along a chosen direction.


```python
# ======================================================
# Function definition
# ======================================================

# Function: f(x,y) = x^2 + y^2 + xy
def f(x, y):
    return x**2 + y**2 + x*y


# ======================================================
# 1. Manual gradient (as in the slides)
# ======================================================

def grad_f(x, y):
    return np.array([2*x + y, x + 2*y])


# ======================================================
# Evaluation point and direction
# ======================================================

point = np.array([1.0, 2.0])

u = np.array([1.0, -1.0])
u = u / np.linalg.norm(u)   # normalize direction


# Manual gradient
g_manual = grad_f(*point)

# Directional derivative
directional_manual = g_manual @ u


# ======================================================
# 2. Symbolic gradient with SymPy
# ======================================================

x, y = sp.symbols('x y')

f_sym = x**2 + y**2 + x*y

grad_sym = sp.Matrix([
    sp.diff(f_sym, x),
    sp.diff(f_sym, y)
])

print("Symbolic gradient:")
sp.pprint(grad_sym)


# Convert symbolic gradient to numerical function
grad_func = sp.lambdify((x, y), grad_sym, "numpy")

g_symbolic = np.array(grad_func(*point), dtype=float).flatten()

directional_symbolic = g_symbolic @ u


# ======================================================
# Results
# ======================================================

print("\nPoint:", point)
print("Direction u:", u)

print("\nManual gradient:", g_manual)
print("Symbolic gradient:", g_symbolic)

print("\nManual directional derivative:", directional_manual)
print("Symbolic directional derivative:", directional_symbolic)
```

### 16. Chain rule

The chain rule differentiates compositions.

If $y = f(u)$ and $u = g(x)$, then

$ \displaystyle \frac{dy}{dx} = \frac{dy}{du}\frac{du}{dx} $.

In deep learning, backpropagation is repeated application of the chain rule.

**Example**

Consider

$ f(u) = \ln(u) $
and
$ g(x) = x^2 + 3x + 1 $.

Then the composition is

$ y(x) = f(g(x)) = \ln(x^2 + 3x + 1) $.

By the chain rule,

$ y'(x) = f'(g(x))\,g'(x) $.

Since

$ \displaystyle f'(u) = \frac{1}{u} $ 
and 
$ g'(x) = 2x + 3 $,

we obtain

$ \displaystyle y'(x) = \frac{1}{x^2 + 3x + 1}(2x + 3) = \frac{2x + 3}{x^2 + 3x + 1} $.


```python
# ======================================================
# Define the functions
# ======================================================

def g(x):
    return x**2 + 3*x + 1

def f(u):
    return np.log(u)

def composed(x):
    return f(g(x))


# ======================================================
# 1. Derivative using the chain rule (manual)
# ======================================================

x_val = 1.0

chain_rule_value = (1 / g(x_val)) * (2*x_val + 3)

print("Manual chain rule derivative:", chain_rule_value)


# ======================================================
# 2. Symbolic computation with SymPy
# ======================================================

x = sp.symbols('x')

g_sym = x**2 + 3*x + 1
f_sym = sp.log(g_sym)

df_sym = sp.diff(f_sym, x)
df_sym = sp.simplify(df_sym)

print("\nSymbolic derivative:")
sp.pprint(df_sym)


# ======================================================
# 3. Evaluate symbolic derivative
# ======================================================

df_func = sp.lambdify(x, df_sym, "numpy")

symbolic_value = df_func(x_val)

print("\nSymbolic derivative at x =", x_val, ":", symbolic_value)


# ======================================================
# 4. Compare both results
# ======================================================

print("\nDifference:", abs(symbolic_value - chain_rule_value))
```

### 17. Jacobian

The **Jacobian matrix** is the matrix of first derivatives of a vector-valued function.

It describes the best local linear approximation of that function.


```python
# ======================================================
# Vector function
# ======================================================

def F(x, y):
    return np.array([x**2 * y, np.sin(x*y)])

# Manual Jacobian (from the slides)
def jacobian_F(x, y):
    return np.array([
        [2*x*y, x**2],
        [y*np.cos(x*y), x*np.cos(x*y)]
    ])


# ======================================================
# Evaluation point
# ======================================================

x_val, y_val = 1.0, 2.0

print("Vector function value:")
print("F(x,y) =", F(x_val, y_val))

print("\nManual Jacobian:")
print(jacobian_F(x_val, y_val))


# ======================================================
# Symbolic Jacobian with SymPy
# ======================================================

x, y = sp.symbols('x y')

F_sym = sp.Matrix([
    x**2 * y,
    sp.sin(x*y)
])

J_sym = F_sym.jacobian([x, y])

print("\nSymbolic Jacobian:")
sp.pprint(J_sym)


# Convert symbolic Jacobian to numerical function
J_func = sp.lambdify((x, y), J_sym, "numpy")

J_symbolic = np.array(J_func(x_val, y_val), dtype=float)

print("\nSymbolic Jacobian evaluated at the point:")
print(J_symbolic)


# ======================================================
# Comparison
# ======================================================

print("\nDifference between manual and symbolic Jacobian:")
print(jacobian_F(x_val, y_val) - J_symbolic)
```

### 18. Hessian

The **Hessian** is the matrix of second derivatives of a scalar function.

It describes curvature and is very important in optimization.


```python
# ======================================================
# Function
# ======================================================

# f(x,y,z) = x^2 + y^2 + z^2 + xy + xz

# Manual Hessian (from the lecture slides)
def hessian_example():
    return np.array([
        [2, 1, 1],
        [1, 2, 0],
        [1, 0, 2]
    ])

H_manual = hessian_example()

print("Manual Hessian:")
print(H_manual)

print("\nIs the Hessian symmetric?")
print(np.allclose(H_manual, H_manual.T))


# ======================================================
# Symbolic computation with SymPy
# ======================================================

x, y, z = sp.symbols('x y z')

f = x**2 + y**2 + z**2 + x*y + x*z

# Compute Hessian symbolically
H_sym = sp.hessian(f, (x, y, z))

print("\nSymbolic Hessian:")
sp.pprint(H_sym)


# Convert symbolic matrix to numerical numpy array
H_symbolic = np.array(H_sym, dtype=float)

print("\nSymbolic Hessian as numpy array:")
print(H_symbolic)


# ======================================================
# Comparison
# ======================================================

print("\nDifference between manual and symbolic Hessian:")
print(H_manual - H_symbolic)
```

### 19. Taylor approximation

Taylor polynomials approximate a function locally around a point.

For example, around $x_0=1$,

$ e^{5(x-1)} \approx 1 + 5(x-1) + \frac{25}{2}(x-1)^2 + \frac{125}{6}(x-1)^3 $.


```python
# ======================================================
# Function
# ======================================================
def f(x):
    return np.exp(5*(x-1))


# ======================================================
# 1. Manual Taylor polynomial (order 3 around x0 = 1)
# ======================================================
def P3_manual(x):
    return 1 + 5*(x-1) + (25/2)*(x-1)**2 + (125/6)*(x-1)**3


# ======================================================
# 2. Symbolic Taylor polynomial with SymPy
# ======================================================

x = sp.symbols('x')

f_sym = sp.exp(5*(x-1))

P3_sym = sp.series(f_sym, x, 1, 4).removeO()

print("Symbolic Taylor polynomial:")
sp.pprint(P3_sym)


# Convert symbolic polynomial to numerical function
P3_sym_func = sp.lambdify(x, P3_sym, "numpy")


# ======================================================
# 3. Evaluation point
# ======================================================

x_val = 1.1

true_value = f(x_val)

manual_value = P3_manual(x_val)

symbolic_value = P3_sym_func(x_val)


# ======================================================
# 4. Results
# ======================================================

print("\nEvaluation at x =", x_val)

print("\nTrue value:")
print(true_value)

print("\nManual Taylor approximation:")
print(manual_value)

print("\nSymbolic Taylor approximation:")
print(symbolic_value)

print("\nManual absolute error:")
print(abs(true_value - manual_value))

print("\nSymbolic absolute error:")
print(abs(true_value - symbolic_value))
```

### 20. Convexity

A function is **convex** if the segment joining two points on its graph lies above the graph.

A key optimization fact is:

- for a convex function, every local minimum is also a global minimum.


```python
def f(x):
    return x**2

x = -1.0
y = 2.0
lam = 0.3

left = f(lam*x + (1-lam)*y)
right = lam*f(x) + (1-lam)*f(y)

print("f(lambda x + (1-lambda) y) =", left)
print("lambda f(x) + (1-lambda) f(y) =", right)
print("Convexity inequality satisfied?", left <= right)
```
